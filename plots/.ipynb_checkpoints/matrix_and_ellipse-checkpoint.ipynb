```python
import matplotlib.pyplot as plt
import numpy as np
```

### HW2, Q5
Consider the matrix
$$
A = \begin{bmatrix}
-2, & 11 \\
-10, & 5
\end{bmatrix}
$$
List the singular values, left singular vectors, and right singular vectors of $A$. Draw a careful, labeled picture of the unit ball in $\mathbb R^2$ and its image under $A$, together with the singular vectors, with the coordinates of their vertices marked.

```python
x = np.zeros(200)
y = np.zeros(200)
x2 = np.zeros(200)
y2 = np.zeros(200)

for i in range(200):
    a = i * 2 * np.pi / 200
    x[i] = np.cos(a)
    y[i] = np.sin(a)
    x2[i] = -2*x[i] + 11*y[i]
    y2[i] = -10*x[i] + 5 * y[i]
```

```python
f, axs = plt.subplots(1,1,figsize=(5,5))
axs.plot(x,y,'.')
axs.plot([x[50], 0,x[0]], [y[50], 0,y[0]])
```

```python
f, axs = plt.subplots(1,1,figsize=(5,5))
axs.plot(x2,y2,'.')
axs.plot([x2[50], 0,x2[0]], [y2[50], 0,y2[0]])  # a1, a2
axs.plot([-5, 0, 10], [5, 0, 10])  #  sigma*u1, sigma*u2
axs.plot([x2[50], 0,x2[0]], [y2[50], 0,y2[0]])  # Av1, Av2
```
